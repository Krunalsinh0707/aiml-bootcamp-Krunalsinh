> **Vectorisation** — performing an operation on an entire array at once, instead of looping over
> elements one at a time. The loop still happens; it happens down in compiled C rather than in Python,
> without interpreter overhead on every iteration.

Let's measure rather than take my word for it.

````python
import timeit

N = 1_000_000
setup = f"import numpy as np; lst = list(range({N})); arr = np.arange({N})"

py_time = timeit.timeit("[x * 2 for x in lst]", setup=setup, number=10) / 10
np_time = timeit.timeit("arr * 2", setup=setup, number=10) / 10

print(f"Python list comprehension: {py_time * 1000:7.2f} ms")
print(f"NumPy vectorised:          {np_time * 1000:7.2f} ms")
print(f"NumPy is about {py_time / np_time:.0f}x faster")
````

Expect roughly 30–60x depending on your machine. You will sometimes see "100x" quoted — achievable on
some operations and hardware, but not a promise. The honest claim: **vectorised NumPy is dramatically
faster, and the gap grows with size.**

<div style="background:#1A1D1D; padding:12px 18px; border-radius:5px;">
<span style="font-family:Calibri,sans-serif; font-size:11px; color:#F05E22; font-weight:bold;
letter-spacing:1px;">PREDICT FIRST</span>
<p style="font-family:Calibri,sans-serif; font-size:13px; color:#fff; margin:5px 0 0 0;">We're about to time the same operation at 1,000 / 10,000 / 100,000 / 1,000,000 elements. Will the speedup be roughly constant, or will it change with size?</p>
</div>

````python
import numpy as np
import timeit

for n in [1_000, 10_000, 100_000, 1_000_000]:
    setup = f"import numpy as np; lst = list(range({n})); arr = np.arange({n})"
    py = timeit.timeit("[x * 2 for x in lst]", setup=setup, number=20) / 20
    npt = timeit.timeit("arr * 2", setup=setup, number=20) / 20
    print(f"n = {n:>9,}   python {py*1000:7.2f} ms   numpy {npt*1000:6.3f} ms   {py/npt:5.0f}x")
````

At small sizes the advantage is modest — NumPy pays a fixed setup cost per operation. At a million
elements it is enormous. **Vectorise when data is big; do not convert a 5-element list to an array.**

> **ndarray** — NumPy's core type: an N-dimensional array whose elements are all the *same* type,
> stored in one contiguous block of memory. That uniformity is what makes it fast.

````python
import numpy as np

print("from a list:      ", np.array([1, 2, 3, 4, 5]))
print("arange(0, 10, 2): ", np.arange(0, 10, 2))     # start, stop, step
print("linspace(0, 1, 5):", np.linspace(0, 1, 5))    # start, stop, how many
print("zeros(5):         ", np.zeros(5))
print("ones(3):          ", np.ones(3))
print("full(4, 7):       ", np.full(4, 7))
````

**`arange` vs `linspace`** — constantly confused:
- `arange(start, stop, step)` — you control the **step size**; stop is **excluded**
- `linspace(start, stop, num)` — you control **how many points**; stop is **included**

````python
import numpy as np

print("arange(0, 1, 0.25):", np.arange(0, 1, 0.25))   # no 1.0 -- stop excluded
print("linspace(0, 1, 5): ", np.linspace(0, 1, 5))    # has 1.0 -- stop included
````

<div style="background:#1A1D1D; padding:12px 18px; border-radius:5px;">
<span style="font-family:Calibri,sans-serif; font-size:11px; color:#F05E22; font-weight:bold;
letter-spacing:1px;">PREDICT FIRST</span>
<p style="font-family:Calibri,sans-serif; font-size:13px; color:#fff; margin:5px 0 0 0;">We slice an array, change the slice, then print the ORIGINAL. Does the original change?</p>
</div>

````python
import numpy as np

original = np.array([1, 2, 3, 4, 5])
piece = original[1:4]
piece[0] = 999

print("piece   :", piece)
print("original:", original, "  <-- CHANGED. slices share memory")
````

````python
import numpy as np

# .copy() gives you independence
original = np.array([1, 2, 3, 4, 5])
safe = original[1:4].copy()
safe[0] = 999

print("safe    :", safe)
print("original:", original, "  <-- untouched")

# contrast with a Python list, which copies on slice
py = [1, 2, 3, 4, 5]
py_piece = py[1:4]
py_piece[0] = 999
print("\npython list original:", py, "  <-- lists COPY on slice")
````

This is a **major** difference from Python lists. NumPy slices are *views* — fast and memory-efficient,
but they will bite you if unexpected. It's the array version of Day 2's mutability trap.

**Rule of thumb:** if you're going to modify a slice and need the original intact, `.copy()`.

> **Boolean mask** — an array of `True`/`False` the same shape as your data, used to select elements.

````python
import numpy as np

scores = np.array([45, 88, 32, 91, 67, 40, 78])

mask = scores >= 60          # a comparison produces booleans
print("scores:", scores)
print("mask:  ", mask)
print("\nselected:", scores[mask])
print("one-liner:", scores[scores >= 60])
````

````python
import numpy as np

scores = np.array([45, 88, 32, 91, 67, 40, 78])

# combine with & (and), | (or), ~ (not) -- parentheses REQUIRED
print("between 40 and 80:", scores[(scores >= 40) & (scores <= 80)])
print("below 40 or above 85:", scores[(scores < 40) | (scores > 85)])
print("NOT passing:", scores[~(scores >= 60)])
````

Two things to burn in:

1. **Use `&`, `|`, `~` — never `and`, `or`, `not`.** The Python keywords try to reduce a whole array to
   one True/False and fail.
2. **Parentheses are mandatory.** `scores >= 40 & scores <= 80` breaks because `&` binds tighter
   than `>=`.

````python
import numpy as np

scores = np.array([45, 88, 32, 91, 67, 40, 78])

# the error you get with 'and'
try:
    scores[(scores >= 40) and (scores <= 80)]
except ValueError as e:
    print("ValueError:", str(e)[:70], "...")
````

That trick — `.sum()` to **count** and `.mean()` to get a **proportion** — you will use constantly.

> **axis** — which dimension to *collapse*. `axis=0` collapses down the rows (giving one value per
> **column**). `axis=1` collapses across the columns (one value per **row**).

The wording trips everyone. The reliable way to remember: **the axis you name is the one that
disappears.**

<div style="background:#1A1D1D; padding:12px 18px; border-radius:5px;">
<span style="font-family:Calibri,sans-serif; font-size:11px; color:#F05E22; font-weight:bold;
letter-spacing:1px;">PREDICT FIRST</span>
<p style="font-family:Calibri,sans-serif; font-size:13px; color:#fff; margin:5px 0 0 0;">grid is shape (2, 3). What shape does grid.sum(axis=0) have — and does it give 2 numbers or 3?</p>
</div>

````python
import numpy as np

grid = np.array([[1, 2, 3],
                 [4, 5, 6]])

col_sums = grid.sum(axis=0)     # axis 0 (size 2) disappears -> 3 left
row_sums = grid.sum(axis=1)     # axis 1 (size 3) disappears -> 2 left

print("grid.sum(axis=0) =", col_sums, " shape", col_sums.shape, " <- per column")
print("grid.sum(axis=1) =", row_sums, " shape", row_sums.shape, " <- per row")
````

`(2, 3)` with `axis=0` → the 2 vanishes → shape `(3,)`. Check the shape of the result and you can
never get this wrong.

````python
import numpy as np

# same idea, real data: 4 students x 3 exams
marks = np.array([[80, 70, 90],
                  [60, 65, 70],
                  [95, 88, 92],
                  [50, 45, 60]])

print("per-student average (axis=1):", marks.mean(axis=1))
print("per-exam average    (axis=0):", marks.mean(axis=0))
````

> **Broadcasting** — NumPy's rules for combining arrays of *different shapes*, by automatically
> stretching the smaller across the larger without copying memory.

You've already used it without noticing.

````python
import numpy as np

a = np.array([1, 2, 3, 4])

# 10 is a single number -- stretched across all four elements
print("a * 10  =", a * 10)
print("a + 100 =", a + 100)
````

### The rule

Compare shapes **from the right**. Two dimensions are compatible if they are **equal**, or one of them
is **1**. Otherwise: error.

```
    (3, 4)          (3, 4)          (3, 4)
    (   4)   OK     (3, 1)   OK     (   3)   ERROR
   -------         -------         -------
    (3, 4)          (3, 4)          4 vs 3 -- neither equal nor 1
```

````python
import numpy as np

grid = np.array([[1, 2, 3],
                 [4, 5, 6]])          # (2, 3)

row = np.array([10, 20, 30])          # (3,) -> stretched DOWN the rows
print("grid + row:")
print(grid + row)
````

````python
import numpy as np

grid = np.array([[1, 2, 3],
                 [4, 5, 6]])          # (2, 3)

col = np.array([[100],
                [200]])               # (2, 1) -> stretched ACROSS the columns
print("grid + col:")
print(grid + col)
````

<div style="background:#1A1D1D; padding:12px 18px; border-radius:5px;">
<span style="font-family:Calibri,sans-serif; font-size:11px; color:#F05E22; font-weight:bold;
letter-spacing:1px;">PREDICT FIRST</span>
<p style="font-family:Calibri,sans-serif; font-size:13px; color:#fff; margin:5px 0 0 0;">grid is (2,3). We add an array of shape (2,). Compatible or error? Compare from the right.</p>
</div>

````python
import numpy as np

grid = np.array([[1, 2, 3],
                 [4, 5, 6]])          # (2, 3)
bad = np.array([1, 2])                # (2,) -> compares 3 vs 2 -- FAILS

try:
    grid + bad
except ValueError as e:
    print("ValueError:", e)
````

````python
import numpy as np

grid = np.array([[1, 2, 3],
                 [4, 5, 6]])
bad = np.array([1, 2])

# fix: reshape (2,) into (2,1) so it stretches across columns
fixed = bad.reshape(2, 1)
print("reshaped to:", fixed.shape)
print(grid + fixed)
````

### A real use: standardising features

Broadcasting doing actual ML work — centring each column by subtracting its mean.

````python
import numpy as np

# 4 samples, 3 features -- note the wildly different scales
data = np.array([[10.0, 200.0, 3.0],
                 [20.0, 300.0, 5.0],
                 [30.0, 100.0, 7.0],
                 [40.0, 400.0, 9.0]])

col_means = data.mean(axis=0)     # (3,) -- one per feature
col_stds = data.std(axis=0)

print("data shape: ", data.shape)
print("means shape:", col_means.shape, "->", col_means)

standardised = (data - col_means) / col_stds     # (4,3) - (3,) broadcasts

print("\nstandardised:")
print(np.round(standardised, 3))
print("\ncolumn means now:", np.round(standardised.mean(axis=0), 10))
print("column stds now: ", np.round(standardised.std(axis=0), 10))
````

That's **feature standardisation** — done before nearly every model in Phase 2. Three lines, no loops,
works on four rows or four million.

````python
import numpy as np

# another classic: a multiplication table, no loops
rows = np.arange(1, 6).reshape(-1, 1)    # (5, 1)
cols = np.arange(1, 6)                   # (5,)

print("(5,1) * (5,) broadcasts to (5,5):")
print(rows * cols)
````

## Putting it together

One realistic problem using everything above — and not a single Python loop.

````python
import numpy as np

rng = np.random.default_rng(7)      # seeded -> reproducible

students = np.array(["Asha", "Ravi", "Meera", "Dev", "Priya"])
marks = rng.integers(35, 100, size=(5, 4))       # 5 students x 4 exams

print("marks:")
print(marks)

per_student = marks.mean(axis=1)                 # collapse exams
per_exam = marks.mean(axis=0)                    # collapse students

print("\nper-student average:", np.round(per_student, 1))
print("per-exam average:   ", np.round(per_exam, 1))

passed = per_student >= 60                       # boolean mask
print("\npassed:", students[passed])
print("failed:", students[~passed])

best = per_student.argmax()                      # argmax gives the INDEX
print(f"\ntop student: {students[best]} ({per_student[best]:.1f})")
print(f"hardest exam: exam {per_exam.argmin() + 1} (avg {per_exam.min():.1f})")
````

````python
import numpy as np

rng = np.random.default_rng(7)
marks = rng.integers(35, 100, size=(5, 4))

# curve everyone +5, cap at 100 -- broadcasting + clip
curved = np.clip(marks + 5, 0, 100)
capped = ((marks + 5) > 100).sum()

print("curved (+5, capped at 100):")
print(curved)
print(f"\nmarks that hit the cap: {capped}")
````

## Recap

| Idea | Syntax | Why it matters |
|---|---|---|
| Create | `np.array`, `arange`, `linspace`, `zeros` | Getting data in |
| Inspect | `.shape`, `.dtype`, `.ndim` | Every bug starts here |
| Index 2D | `arr[row, col]`, `arr[:, 1]` | Rows and columns |
| Views | `.copy()` when you need independence | Silent bugs otherwise |
| Mask | `arr[arr > 5]`, `&`, `\|`, `~` | Filtering, no loops |
| Count | `mask.sum()`, `mask.mean()` | True == 1 |
| Conditional | `np.where(cond, a, b)` | Vectorised if/else |
| Aggregate | `.mean(axis=0)` vs `axis=1` | The named axis disappears |
| Broadcasting | shapes align from the right | The core idea |
| Reshape | `.reshape(-1, 1)` | Making shapes fit |

### The two things to remember

1. **Never write a Python loop over an array.** If you're looping, there's a vectorised way — and it
   will be dramatically faster.
2. **When something breaks, print the shapes first.** Nine out of ten NumPy errors are shape errors.
